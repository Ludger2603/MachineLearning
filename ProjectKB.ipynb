```python
# modules we'll use
import pandas as pd
import numpy as np

# for Box-Cox Transformation
from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

# read in all our data
kickstarters_2017 = pd.read_csv("/content/tingkat-pengangguran-jakarta.csv")

# set seed for reproducibility
np.random.seed(0)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.preprocessing import minmax_scaling

# generate 1000 data points randomly drawn from an exponential distribution
original_data = np.random.exponential(size=1000)

# min-max scale the data between 0 and 1
scaled_data = minmax_scaling(original_data, columns=[0])

# plot both together to compare
fig, ax = plt.subplots(1, 2)
sns.histplot(original_data, ax=ax[0], kde=True)
ax[0].set_title("Original Data")
sns.histplot(scaled_data, ax=ax[1], kde=True)
ax[1].set_title("Scaled data")

plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.preprocessing import minmax_scaling

# Select the jumlah_pengangguran_terbuka column
jumlah_pengangguran = kickstarters_2017['jumlah_pengangguran_terbuka']

# Convert jumlah_pengangguran to a NumPy array
jumlah_pengangguran_array = jumlah_pengangguran.values

# Scale the data from 0 to 1
scaled_data = minmax_scaling(jumlah_pengangguran_array, columns=[0])

# Plot the original & scaled data together to compare
fig, ax = plt.subplots(1, 2)
sns.histplot(jumlah_pengangguran, ax=ax[0], kde=True)
ax[0].set_title("Original Data")
sns.histplot(scaled_data, ax=ax[1], kde=True)
ax[1].set_title("Scaled data")

plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats  # Import modul stats dari scipy

# get the index of all positive pledges
index_of_positive_pledges = kickstarters_2017['jumlah_pengangguran_terbuka'] > 0

# get only positive pledges
positive_pledges = kickstarters_2017['jumlah_pengangguran_terbuka'].loc[index_of_positive_pledges]

# normalize the pledges (w/ Box-Cox)
normalized_pledges = stats.boxcox(positive_pledges + 1)[0]  # Tambahkan 1 untuk menghindari nilai 0

# plot both together to compare
fig, ax = plt.subplots(1, 2)
sns.histplot(positive_pledges, ax=ax[0], kde=True)
ax[0].set_title("Original Data")
sns.histplot(normalized_pledges, ax=ax[1], kde=True)
ax[1].set_title("Normalized data")

plt.show()
```
